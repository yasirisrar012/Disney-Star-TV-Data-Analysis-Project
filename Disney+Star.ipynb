# Step-3 : Data Preprocessing and Manipulation

```python
On_air_channels = pd.read_csv('On_air_channels.csv')
On_air_channels
```

```python
import numpy as np
```

```python
On_air_channels['Average Daily Viewership (millions)'] = np.round(np.random.uniform(0.5, 10, size=len(On_air_channels)), 2)
On_air_channels['Monthly Active Viewers (millions)'] = np.round(np.random.uniform(1, 50, size=len(On_air_channels)), 2)
On_air_channels['Peak Viewership (millions)'] = np.round(np.random.uniform(5, 20, size=len(On_air_channels)), 2)

On_air_channels['Advertising Revenue (million USD)'] = np.round(np.random.uniform(10, 500, size=len(On_air_channels)), 2)
On_air_channels['Subscription Revenue (million USD)'] = np.round(np.random.uniform(5, 300, size=len(On_air_channels)), 2)
On_air_channels['Total Annual Revenue (million USD)'] = On_air_channels['Advertising Revenue (million USD)'] + On_air_channels['Subscription Revenue (million USD)']

On_air_channels['Average Daily Broadcast Hours'] = np.random.randint(10, 24, size=len(On_air_channels))
On_air_channels['Prime-Time Broadcast Hours per Day'] = np.random.randint(2, 8, size=len(On_air_channels))

On_air_channels['Total Subscribers (millions)'] = np.round(np.random.uniform(1, 200, size=len(On_air_channels)), 2)
On_air_channels['Number of Unique Programs/Shows'] = np.random.randint(5, 50, size=len(On_air_channels))
On_air_channels['Number of Original Productions'] = np.random.randint(1, 20, size=len(On_air_channels))

On_air_channels['Average Content Rating (1-10)'] = np.round(np.random.uniform(5, 10, size=len(On_air_channels)), 1)
On_air_channels['Annual Growth Rate of Viewership (%)'] = np.round(np.random.uniform(-5, 20, size=len(On_air_channels)), 2)
```
